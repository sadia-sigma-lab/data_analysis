```python
  
import json
import os 
import glob
import pandas  as pd

path_to_json = r'/content/drive/MyDrive/Jan/data_engineer_case_study/data/readings/'
# import all files from folder which ends with .json 
json_files = glob.glob(os.path.join(path_to_json, '*.json'))
print(json_files)
df = pd.DataFrame( columns=['interval_start' , 'consumption_delta' , 'meterpoint_id'])
for file in json_files:
# Opening JSON file
  f = open(file)
    
  # returns JSON object as 
  # a dictionary
  data = json.load(f)
  # print(data)

  for i in data['data']:
    #   append all json data into dataframe
    df.loc[len(df.index)] = i
  df.head()
  
```

```python
import pandas as pd
import sqlite3
table_name = 'json_data'
conn = sqlite3.connect('mydb.sqlite')
query = f'Create table if not Exists {table_name}  (interval_start text, consumption_delta real, meterpoint_id string)'
conn.execute(query)
df.to_sql(table_name,conn,if_exists='replace',index=False)
conn.commit()

r_df = pd.read_sql("select * from json_data",conn)
print(r_df)
```
